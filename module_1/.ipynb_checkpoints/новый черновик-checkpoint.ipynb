# Предобработка

```python
answers = {} # создадим словарь для ответов

# тут другие ваши предобработки колонок например:

#the time given in the dataset is in string format.
#So we need to change this in datetime format
# ...


# подработочные словари для столбцов со множественными значениями
genres = [] # список встречающихся жанров         пока словарь
cast = [] # список актеров снимавшихся в выборке  пока словарь
director = [] # список режисеров                  пока словарь
production_companies = [] # списоккинокомпаний    пока словарь

def genres_i():
    # Создаем список встречающихся жанров
    for i in data['genres'].unique():
        ii = i.split("|")
        for j in ii:
            if not (j in genres):
                genres.append(j)
    return genres

def cast_i():
    # Создаем список встречающихся жанров
    for i in data['cast'].unique():
        ii = i.split("|")
        for j in ii:
            if not (j in cast):
                cast.append(j)
    return cast

def director_i():
    for i in data['director'].unique():
        ii = i.split("|")
        for j in ii:
            if not (j in director):
                director.append(j)
    return director

def production_companies_i():
    for i in data['production_companies'].unique():
        ii = i.split("|")
        for j in ii:
            if not (j in production_companies):
                production_companies.append(j)
    return production_companies

                
production_companies
```

# 6. Какой самый прибыльный фильм?
#### Внимание! Здесь и далее под «прибылью» или «убытками» понимается разность между сборами и бюджетом фильма. (прибыль = сборы - бюджет) в нашем датасете это будет (profit = revenue - budget) 

```python
# лучше код получения столбца profit вынести в Предобработку что в начале
res = data
res['Profit'] = res['revenue'] - res['budget'] # Добавили подработачный столбец Profit

# res[res['Profit'] == res['Profit'].max()][['original_title', 'imdb_id']]
res[['original_title', 'imdb_id', 'Profit']].sort_values(['Profit'],  ascending=False).head()
# res.head()
```

```python
genres_s = {}
# Создаем словарь встречающихся жанров, где значение ключа - число в датасете данного ключа
for i in data['genres'].unique():
    ii = i.split("|")
    for j in ii:
        genres_s.setdefault(j, 1)
        if genres_s[j] != 0:
            genres_s[j] += 1


dict(sorted(genres_s.items(), key=lambda x: x[1], reverse = True))
```

# 12. Фильмы какого жанра чаще всего становятся прибыльными? 

```python
df = data
# Создаем новый столбец содержащий список значений (парсинг строки)
df['genres1'] = df['genres'].str.split('|')

# df['genres1'].apply(pd.Series).stack().reset_index(1, drop=True)

# df
dummies_df = pd.get_dummies(
  df.join(pd.Series(df['genres1'].apply(pd.Series).stack().reset_index(1, drop=True), 
                    name='genres2')).drop('genres1', axis=1).rename(columns={'genres2': 'genres1'}),
                    columns=['genres1']).groupby('original_title', as_index=False).sum()

dummies_df
```
